```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os 
from matplotlib.lines import Line2D
import matplotlib.patches as mpatches

folder_figures = 'figures'
```

```python
# Results from true tops
metric = 'spearman'

fp_databse = "data/database_640.pkl"
database = pd.read_pickle(fp_databse)
true_tm = database['tm'].to_list()

folder_pred = 'data/predictions_exp_blosum_640_mae_no_mean'
folder_perf = 'data/performances_exp_blosum_640_mae_no_mean'

fp_pred_ave_true_top = os.path.join(folder_pred, f'ens_no_mean_{metric}_ave_true_top.csv')
fp_pred_ave_div_emb = os.path.join(folder_pred, f'ens_no_mean_{metric}_ave_div_emb.csv')
fp_pred_ave_div_reg = os.path.join(folder_pred, f'ens_no_mean_{metric}_ave_div_reg.csv')

fp_pred_ridge = os.path.join(folder_pred, f'ens_no_mean_{metric}_true_top.csv')
fp_pred_ridge_topemb = os.path.join(folder_pred, f'ens_no_mean_{metric}_div_emb.csv')
fp_pred_ridge_topreg = os.path.join(folder_pred, f'ens_no_mean_{metric}_div_reg.csv')

df_pred_ave_true_top = pd.read_csv(fp_pred_ave_true_top)
df_pred_ave_div_emb = pd.read_csv(fp_pred_ave_div_emb)
df_pred_ave_div_reg = pd.read_csv(fp_pred_ave_div_reg)

df_pred_ridge_top = pd.read_csv(fp_pred_ridge)
df_pred_ridge_topemb = pd.read_csv(fp_pred_ridge_topemb)
df_pred_ridge_topreg = pd.read_csv(fp_pred_ridge_topreg)

fp_ave_true_top = os.path.join(folder_perf, f'ens_no_mean_{metric}_ave_true_top.csv')
fp_ave_div_emb = os.path.join(folder_perf, f'ens_no_mean_{metric}_ave_div_emb.csv')
fp_ave_div_reg = os.path.join(folder_perf, f'ens_no_mean_{metric}_ave_div_reg.csv')

fp_ridge_metrics = os.path.join(folder_perf, f'ens_no_mean_{metric}_true_top.csv')
fp_ridge_topemb = os.path.join(folder_perf, f'ens_no_mean_{metric}_div_emb.csv')
fp_ridge_topreg = os.path.join(folder_perf, f'ens_no_mean_{metric}_div_reg.csv')


ave_true_top = pd.read_csv(fp_ave_true_top)
ave_true_topemb = pd.read_csv(fp_ave_div_emb)
ave_true_topreg = pd.read_csv(fp_ave_div_reg)

ridge_metrics_top = pd.read_csv(fp_ridge_metrics)
ridge_metrics_topemb = pd.read_csv(fp_ridge_topemb)
ridge_metrics_topreg = pd.read_csv(fp_ridge_topreg)
```

```python
max_tm = 8 

# AVG
ave_spearmans_top = np.array(ave_true_top['spearman'].to_list()[:max_tm])
ave_spearmans_std_top = np.array(ave_true_top['std_spearman'].to_list()[:max_tm])
ave_sdr_top = np.array(ave_true_top['sdr'].to_list()[:max_tm])
ave_sdr_std_top = np.array(ave_true_top['std_sdr'].to_list()[:max_tm])

ave_spearmans_topemb = np.array(ave_true_topemb['spearman'].to_list()[:max_tm])
ave_spearmans_std_topemb = np.array(ave_true_topemb['std_spearman'].to_list()[:max_tm])
ave_sdr_topemb = np.array(ave_true_topemb['sdr'].to_list()[:max_tm])
ave_sdr_std_topemb = np.array(ave_true_topemb['std_sdr'].to_list()[:max_tm])

ave_spearmans_topreg = np.array(ave_true_topreg['spearman'].to_list()[:max_tm]  + [np.nan])
ave_spearmans_std_topreg = np.array(ave_true_topreg['std_spearman'].to_list()[:max_tm] + [np.nan])
ave_sdr_topreg = np.array(ave_true_topreg['sdr'].to_list()[:max_tm] + [np.nan])
ave_sdr_std_topreg = np.array(ave_true_topreg['std_sdr'].to_list()[:max_tm] + [np.nan]) 

# RIDGE STACKING
ridge_spearmans_top = np.array(ridge_metrics_top['fin_spearman_test'].to_list()[:max_tm])
ridge_spearmans_std_top = np.array(ridge_metrics_top['std_spearman_test'].to_list()[:max_tm])
ridge_sdr_top = np.array(ridge_metrics_top['fin_std_ratio_test'].to_list()[:max_tm])
ridge_sdr_std_top = np.array(ridge_metrics_top['std_std_ratio_test'].to_list()[:max_tm])

ridge_spearmans_topemb = np.array(ridge_metrics_topemb['fin_spearman_test'].to_list()[:max_tm])
ridge_spearmans_std_topemb = np.array(ridge_metrics_topemb['std_spearman_test'].to_list()[:max_tm])
ridge_sdr_topemb = np.array(ridge_metrics_topemb['fin_std_ratio_test'].to_list()[:max_tm])
ridge_sdr_std_topemb = np.array(ridge_metrics_topemb['std_std_ratio_test'].to_list()[:max_tm])

ridge_spearmans_topreg = np.array(ridge_metrics_topreg['fin_spearman_test'].to_list()[:max_tm] + [np.nan])
ridge_spearmans_std_topreg = np.array(ridge_metrics_topreg['std_spearman_test'].to_list()[:max_tm] + [np.nan])
ridge_sdr_topreg = np.array(ridge_metrics_topreg['fin_std_ratio_test'].to_list()[:max_tm] + [np.nan])
ridge_sdr_std_topreg = np.array(ridge_metrics_topreg['std_std_ratio_test'].to_list()[:max_tm] + [np.nan])
```

```python
# Plot mean vs ridge spearman and std ratio
num_to_plot = 8
num_tm = range(1,num_to_plot+1)

sns.set_style('white', {'axes.spines.right': False, 'axes.spines.top': False, 'axes.spines.bottom': True,
                        'xtick.bottom': True, 'xtick.top': False, 'ytick.left': True, 'xtick.labeltop': False})
fig, ax = plt.subplots(figsize=(12, 7))

# Plotting
linewid = 4

# ax.fill_between(num_tm[:num_to_plot],  ave_spearmans_top[:num_to_plot] + ave_spearmans_std_top[:num_to_plot],  ave_spearmans_top[:num_to_plot] - ave_spearmans_std_top[:num_to_plot], color='darkred', alpha=.08, hatch='\\')
# ax.fill_between(num_tm[:num_to_plot],  ridge_spearmans_top[:num_to_plot] + ridge_spearmans_std_top[:num_to_plot],  ridge_spearmans_top[:num_to_plot] - ridge_spearmans_std_top[:num_to_plot],color='darkred', alpha=.08)

# ax.fill_between(num_tm[:num_to_plot],  ave_spearmans_topemb[:num_to_plot] + ave_spearmans_std_topemb[:num_to_plot],  ave_spearmans_topemb[:num_to_plot] - ave_spearmans_std_topemb[:num_to_plot], color='darkgreen', alpha=.08, hatch='\\')
# ax.fill_between(num_tm[:num_to_plot],  ridge_spearmans_topemb[:num_to_plot] + ridge_spearmans_std_topemb[:num_to_plot],  ridge_spearmans_topemb[:num_to_plot] - ridge_spearmans_std_topemb[:num_to_plot],color='darkgreen', alpha=.08)

# ax.fill_between(num_tm[:num_to_plot],  ave_spearmans_topreg[:num_to_plot] + ave_spearmans_std_topreg[:num_to_plot],  ave_spearmans_topreg[:num_to_plot] - ave_spearmans_std_topreg[:num_to_plot], color='mediumpurple', alpha=.08, hatch='\\')
# ax.fill_between(num_tm[:num_to_plot],  ridge_spearmans_topreg[:num_to_plot] + ridge_spearmans_std_topreg[:num_to_plot],  ridge_spearmans_top[:num_to_plot] - ridge_spearmans_std_topreg[:num_to_plot],color='mediumpurple', alpha=.08)


plt.plot(num_tm[:num_to_plot], ave_spearmans_top[:num_to_plot], color='darkred', linestyle='dotted',linewidth=linewid,alpha=0.75)
plt.plot(num_tm[:num_to_plot], ave_spearmans_topemb[:num_to_plot], color='darkgreen', linestyle='dotted',linewidth=linewid,alpha=0.75)
plt.plot(num_tm[:num_to_plot], ave_spearmans_topreg[:num_to_plot], color='mediumpurple', linestyle='dotted',linewidth=linewid,alpha=0.75)

plt.plot(num_tm[:num_to_plot], ridge_spearmans_top[:num_to_plot], color='darkred',linewidth=linewid,alpha=0.75)
plt.plot(num_tm[:num_to_plot], ridge_spearmans_topemb[:num_to_plot], color='darkgreen',linewidth=linewid,alpha=0.75)
plt.plot(num_tm[:num_to_plot], ridge_spearmans_topreg[:num_to_plot], color='mediumpurple',linewidth=linewid,alpha=0.75)


plt.xlabel(f'',fontsize=30)
plt.ylabel(f'',fontsize=30)


plt.ylim(0.795,0.845)
plt.tick_params(labelsize=22)
plt.xlim(0.5,num_to_plot+0.5)
plt.gca().xaxis.set_major_locator(plt.MultipleLocator(1))
plt.gca().xaxis.set_minor_locator(plt.AutoLocator()) 

handles = [mpatches.Patch(color='darkgreen', label='Top for each embedding', alpha=.75, edgecolor='black'),
mpatches.Patch(color='mediumpurple', label='Top for each regression', alpha=.75, edgecolor='black'),
mpatches.Patch(color='darkred', label='Top-performing combinations', alpha=.75, edgecolor='black'),
            Line2D([0], [0], linestyle='-', color='grey',  linewidth=4, label="Ridge stacking"),
            Line2D([0], [0], linestyle='dotted', color='grey', linewidth=4, label='Averaging'),
            ]

legend = ax.legend(handles=handles,
                    loc='lower right', ncol=1,frameon=True, markerscale=1.6, fontsize=22, edgecolor = 'w', handlelength=1.2, handleheight=0.55)

plt.savefig(os.path.join(folder_figures, f'figure4b.png'), dpi=800, bbox_inches='tight', transparent=True)

plt.show()
plt.close()
```
